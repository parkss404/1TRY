```python
import openpyxl

#한 건의 대화에 대한 정보를 담는 객체이다. 
class Conversation :
    #질문과 응답 두 변수로 구성된다.
    
    #질문
    def __init__(self,contentName,contentType,question,answer):
        self.contentName = contentName
        self.contentType = contentType
        self.question = question
        self.answer = answer
    #응답    
    def __str__(self):
        return "질문: " + self.question + "\n답변: " + self.answer + "\n"
    
#데이터가 담긴 엑셀파일을 열기 
wb = openpyxl.load_workbook("test.xlsx")
ws = wb.active

conversations = []

#데이터를 객체로 가져오기 
for r in ws.rows:
    c = Conversation(r[0].value,r[1].value, r[2].value, r[3].value)
    conversations.append(c)

wb.close()

for c in conversations:
    print(str(c))

print("fdf")

##엑셀 파일에 정수형을 사용하면   print(str(c))이 부분에서
##TypeError: can only concatenate str (not “int”) to str 이 오류가 뜨게 됨.
```
